# Limpeza e otimização dos dados

Adição da coluna Tema, baseado na localização e na referência CDU - Classificação Decimal Universal.

* 000 a 099: Generalidades. Ciência e conhecimento.
* 100 a 199: Filosofia e psicologia.
* 200 a 299: Religião.
* 300 a 399: Ciências sociais.
* 400 a 499: Classe vaga. Provisoriamente não ocupada.
* 500 a 599: Matemática e ciências naturais.
* 600 a 699: Ciências aplicadas.
* 700 a 799: Belas artes.
* 800 a 899: Linguagem. Língua. Linguística.
* 900 a 999: Geografia. Biografia. História."

```python
# Adição de uma coluna tema
tema = []

for row in emprestimos_completo["localizacao"]:
    if row <= 99:
      tema.append("Generalidades. Ciência e conhecimento")
    elif row <= 199:
      tema.append("Filosofia e psicologia")
    elif row <= 299:
      tema.append("Religião")
    elif row <= 399:
      tema.append("Ciências sociais")
    elif row <= 499:
      tema.append("Classe vaga. Provisoriamente não ocupada")
    elif row <= 599:
      tema.append("Matemática e ciências naturais")
    elif row <= 699:
      tema.append("Ciências aplicadas")
    elif row <= 799:
      tema.append("Belas artes")
    elif row <= 899:
      tema.append("Linguagem. Língua. Linguística")
    elif row <= 999:
      tema.append("Geografia. Biografia. História.")

tema = pd.Series(tema)
```

```python
emprestimos_completo["tema"] = tema
```

# Exploração dos dados



**Objetivo:**
A diretoria da biblioteca gostaria de entender se a quantidade de empréstimos está diminuindo, aumentando ou permanecendo igual ao decorrer dos últimos anos.

```python
quantidade_exemplares = emprestimos_completo[["data_emprestimo","id_exemplar"]]
quantidade_exemplares["data_emprestimo"] = quantidade_exemplares["data_emprestimo"].dt.year
quantidade_exemplares.drop_duplicates(inplace=True)

exemplares_por_ano = quantidade_exemplares.value_counts().groupby("data_emprestimo").sum()
exemplares_por_ano
```

### Quantidade de exemplares emprestados por ano

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_axes([0,0,1,1])

ax.plot(exemplares_por_ano)
ax.set_title("Qauntidade de exemplares por ano")
ax.set_ylabel("Quantidade de exemplares")
plt.show()
```

**Análise**:Como é possível observar no gráfico acima, o número de exemplares emprestados era crescente entre 2010-2017, porém teve uma pequena queda até 2019 que pode ser um resultado comum. Mas de 2019 para 2020 houve uma brusca queda que provavelmente ocorreu devido a pandemia de COVID-19. Isso fez com que as pessoas não frequentassem a universidade por conta da doença, acarretando em menor número de empréstimos.

---

**Objetivo 2:** A diretoria também gostaria de gerenciar melhor os recursos humanos da biblioteca de acordo com a demanda de trabalho existente. Por exemplo:


* gerenciar a programação de férias dos colaboradores de acordo com os meses de menor demanda;
programar atividades que não sejam de atendimento ao usuário para períodos específicos de menor demanda.

* Há uma suspeita interna de que os meses com maior número de exemplares emprestados sejam março e setembro, mas não foi realizada uma análise real sobre isso.

```python
quantidade_exemplares = emprestimos_completo[["data_emprestimo","id_exemplar"]]
quantidade_exemplares["data_emprestimo"] = quantidade_exemplares["data_emprestimo"].dt.month
quantidade_exemplares.drop_duplicates(inplace=True)

exemplares_por_mes = quantidade_exemplares.value_counts().groupby("data_emprestimo").sum()

nomes_meses=["Janeiro","Fevereiro","Março","Abril","Maio","Junho","Julho","Agosto","Setembro","Outubro","Novembro","Dezembro"]
exemplares_por_mes = exemplares_por_mes.to_frame("quantidade")
exemplares_por_mes.index = nomes_meses
exemplares_por_mes
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_axes([0,0,1,1])

ax.plot(exemplares_por_mes,color="r")
ax.set_title("Qauntidade de exemplares por ano")
ax.set_ylabel("Quantidade de exemplares")

plt.show()
```

**Análise**: De acordo com o gráfico os meses de menor demanda são Dezembro-Janeiro seguidos de Junho-Julho, o que era de se esperar visto que são os períodos de férias dos alunos, que compôe grande parte do corpo de pessoas que fazem empréstimos na biblioteca. Portanto, o ideal é dar férias aos funcionários prioritariamente dentro desses dois períodos.

Obs: Março e Setembro são uns dos períodos de maior movimento. Por isso é importante fazer a análise dos dados antes de tirar quaisquer conclusões.

---

**Objetivo 3**: Além do gerenciamento anual das atividades, a diretoria também necessita que seja planejada uma programação diária das atividades. Por este motivo, verifique quais foram os horários com maior quantidade de empréstimos ao longo de um dia inteiro.

```python
quantidade_exemplares = emprestimos_completo[["data_emprestimo","id_exemplar"]]
quantidade_exemplares["data_emprestimo"] = quantidade_exemplares["data_emprestimo"].dt.hour
emprestimos_por_hora = quantidade_exemplares.value_counts().groupby("data_emprestimo").sum()

emprestimos_por_hora
```

```python
# Retirando esses 17 empréstimos da meia noite e 50 as 23 horas, pois a biblioteca provavelmente não está aberta.
emprestimos_por_hora = emprestimos_por_hora.iloc[1:-1]
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_axes([0,0,1,1])

bars = ax.bar(x=emprestimos_por_hora.index,height=emprestimos_por_hora.values,tick_label=emprestimos_por_hora.index)
ax.bar_label(bars,label_type="edge")
ax.set_title("Emprestimos realizados por hora")
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
ax.set_ylabel("Exemplares emprestados")
ax.set_xlabel("Horas do dia")
plt.show()
```

**Análise**: Dividindo em dois períodos, temos que na parte da manhã, os horaríos de maior movimento são entre 10-12, enquanto que no período da tarde o maior movimento ocorre entre 16-18. Portanto, esses são os melhores horários para se ter um corpo maior de trabalho na área de atendimento.

```python
def analise_categorica(cat):
  emprestimos_por_cat = emprestimos_completo[[cat,"id_exemplar"]]
  emprestimos_por_cat = emprestimos_por_cat.value_counts().groupby(cat).sum()
  emprestimos_por_cat = emprestimos_por_cat.to_frame("quantidade")
  emprestimos_por_cat.index.name = cat
  emprestimos_por_cat = emprestimos_por_cat.reset_index()
  emprestimos_por_cat["porcentagem"] = round(100*(emprestimos_por_cat["quantidade"]/emprestimos_por_cat["quantidade"].sum()),2)
  return emprestimos_por_cat
```

**Objetivo 4**: Como se distribuem os empréstimos de exemplares pelos tipos de vínculo dos usuários?

```python
emprestimos_por_vinculo = analise_categorica("tipo_vinculo_usuario")
emprestimos_por_vinculo
```

```python
# Gráfico com pelo menos 10000 empréstimos
emprestimos_10000 = emprestimos_por_vinculo.query("quantidade >= 10000")
fig = plt.figure(figsize=(12,8))
ax = sns.barplot(data=emprestimos_10000,y='quantidade',x='tipo_vinculo_usuario',dodge=False)

plt.legend([],[], frameon=False)
ax.set(xlabel=None,ylabel=None)
ax.set_title("Quantidade de emprestimos por vínculo"+"\n",size=20,loc='left',weight='bold')
ax.text(s='Apenas vínculos com mais de 10.000 empréstimos',x=-0.5,y=1700000,fontsize=18, ha='left',color='gray')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
ax.tick_params(axis='x',rotation=30)
```

**Análise**: Fica claro, logo de início que a maioria esmagadora dos usuários da biblioteca são alunos de graduação. O gráfico, feito com vinculos acimade 10.000 empréstimos representa ainda mais essa maioria.

# Análise de emprestimos mensais por ano da coleção acervo ciruclante

Como visto anteriormente, a coleção que mais é emprestada é o arcervo circulante

```python
emprestimo_circulante = emprestimos_completo.query("colecao == 'Acervo Circulante'")
emprestimo_circulante.head()
```

Função para gerar emprestimos mensais por ano

```python
def emprestimos_mensais(ano, vinculo):
  df = emprestimo_circulante.query("tipo_vinculo_usuario == @vinculo")
  df = df.query("data_emprestimo.dt.year == @ano")
  df["data_emprestimo"] = df["data_emprestimo"].dt.month
  df = df[["data_emprestimo","id_exemplar"]].value_counts().groupby("data_emprestimo").sum()
  return df

```

```python
df_graduacao = pd.DataFrame()
df_pos = pd.DataFrame()
for ano in range(2010,2021):
  df_graduacao = pd.concat([df_graduacao,emprestimos_mensais(ano, "ALUNO DE GRADUAÇÃO")], axis=1)
  df_pos = pd.concat([df_pos,emprestimos_mensais(ano, "ALUNO DE PÓS-GRADUAÇÃO")], axis=1)
```

```python
df_graduacao.columns=range(2010,2021)
df_graduacao.index = nomes_meses
df_graduacao
```

```python
plt.figure(figsize=(12,8))
ax = sns.boxplot(data=df_graduacao, palette="rocket")
```

**Análise**: É importante ressaltar que esse recorte, reflete grande parte da estrutura geral do conjunto de dados, pois são os tipos de usuários com maior frequência (alunos de graduação) e da coleção mais utilizada por eles (acervo circulante).

Com o boxplot é possível observar a oscilação da quantidade de empréstimos realizados mensalmente durante cada ano.

Além da análise geral realizada (declínio a partir de 2013, forte baixa em 2019), chama atenção os anos de 2014 e 2017. Que se excluíssemos os outliers (possivelmente meses de janeiro e dezembro) teríamos anos com uma quantidade constante de empréstimos mensais.

Mas se diferenciam principalmente porque em 2014 houve um declínio e em 2017 houve um aumento, em relação aos anos anteriores. Seria interessante entender com a direção das bibliotecas o que ocorreu nessas datas para que pudéssemos ter um contexto melhor da realidade do negócio.

```python
df_pos.columns=range(2010,2021)
df_pos.index = nomes_meses
df_pos
```

```python
plt.figure(figsize=(12,8))
ax = sns.boxplot(data=df_pos,palette="rocket")
```

# Análise de diferença percentual entre 2017 e 2019

## Filtrando tabelas por vinculo

```python
alunos_pos = df_matriculas.query("tipo_vinculo_usuario == 'ALUNO DE PÓS-GRADUAÇÃO'")[["matricula_ou_siape","curso"]]
alunos_pos
```

```python
emprestimos_pos = emprestimos_completo.query("tipo_vinculo_usuario == 'ALUNO DE PÓS-GRADUAÇÃO'")[["matricula_ou_siape","data_emprestimo"]]
emprestimos_pos = emprestimos_pos.query("data_emprestimo > 2017 and data_emprestimo < 2020")
emprestimos_pos
```

```python
emprestimos_partir_2017 = emprestimos_pos.merge(alunos_pos)
emprestimos_partir_2017["data_emprestimo"] = emprestimos_partir_2017["data_emprestimo"].dt.year
emprestimos_partir_2017 = emprestimos_partir_2017[["data_emprestimo","curso"]].value_counts().reset_index()
emprestimos_partir_2017.columns = ["ANO","CURSO","QUANTIDADE DE EMPRESTIMOS"]
emprestimos_partir_2017
```

```python
emprestimos_partir_2017_pivot = emprestimos_partir_2017.pivot_table(index="CURSO",columns="ANO",values="QUANTIDADE DE EMPRESTIMOS", aggfunc=sum)
emprestimos_partir_2017_pivot
```

## Importando previsão de 2022 previamente feita

```python
previsao_2022 = pd.read_table("https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_7-Apresentando_resultados_em_HTML/Dataset/previsao")
previsao_2022
```

```python
previsao_2022 = previsao_2022['curso previsao_2022'].str.split(' ',expand=True)
previsao_2022.columns = ["curso", "quantidade"]
previsao_2022
```

## Tabela pivotada com todos os anos

```python
previsao_2022.index = emprestimos_partir_2017_pivot.index
emprestimos_partir_2017_pivot["2022"] = previsao_2022.iloc[:,1]
emprestimos_partir_2017_pivot
```

```python
emprestimos_partir_2017_pivot["2022"] = emprestimos_partir_2017_pivot["2022"].astype("int")
```

## Tabela de diferenças

```python
def diferenca_ano_anterior(x,y):
  return round(((y-x)/x)*100,2)
```

```python
dif_2018 = diferenca_ano_anterior(emprestimos_partir_2017_pivot.iloc[:,0],emprestimos_partir_2017_pivot.iloc[:,1])
dif_2019 = diferenca_ano_anterior(emprestimos_partir_2017_pivot.iloc[:,1],emprestimos_partir_2017_pivot.iloc[:,2])
dif_2022 = diferenca_ano_anterior(emprestimos_partir_2017_pivot.iloc[:,2],emprestimos_partir_2017_pivot.iloc[:,1])
```

```python
dif_2018
```

```python
diferencas_anos = pd.DataFrame({'2017-2018':dif_2018,
                                '2018-2019':dif_2019,
                                '2019-2022':dif_2022})
diferencas_anos
```

## Formatando a tabela e transformando em HTML

```python
diferencas_anos.reset_index(inplace=True)
diferencas_anos.columns = diferencas_anos.columns.str.capitalize()
diferencas_anos["Curso"] = diferencas_anos["Curso"].str.capitalize()
```

```python
diferencas_anos
```

```python
th_props = [
  ('font-size', '1.4rem'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', 'whitesmoke'),
  ('background-color', '#001692'),
  ('border-radius', '0.25rem'),
  ('box-shadow','0 0 1rem gray')
  ]

td_props = [
  ('font-size', '1rem'),
  ('padding','0.5rem'),
  ('text-align', 'left'),
  ('font-weight', 'bold'),
  ('border-bottom','0.1rem solid lightgray')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]
```

```python
diferencas_anos.style.text_gradient(cmap='RdYlGn',low=1, axis=1,vmax=0.1,vmin=0)\
                                    .hide_index()\
                                    .format("{:.2f} %", subset=["2017-2018","2018-2019","2019-2022"])\
                                    .set_table_styles(styles)\
                                    .to_html("Tabela.html",table_attributes='ALIGN=LEFT WIDTH=50% CELLSPACING = 5')
```
