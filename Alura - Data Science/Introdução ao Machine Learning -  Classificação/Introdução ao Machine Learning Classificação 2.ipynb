```python
import pandas as pd

df = pd.read_csv("https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv")
df.head(5)
```

```python
X = df[["home","how_it_works","contact", "bought"]]
y = df["bought"]

X.head()
```

```python
y.head()
```

```python
df.shape
```

```python
X_train = X[:75]
X_test = X[75:]
y_train = y[:75]
y_test = y[75:]

print(f"Treinaremos o modelo com {len(X_train)} elementos e faremos o teste com {len(X_test)} elementos")
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(X_train, y_train)
previsoes = modelo.predict(X_test)

acuracia = accuracy_score(y_test, previsoes)
print(f"A acuracia foi {acuracia*100:.2f} %")
```

# Usando SKLearn para separar casos de treino e teste

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25,random_state=20)

modelo = LinearSVC()
modelo.fit(X_train, y_train)
previsoes = modelo.predict(X_test)

acuracia = accuracy_score(y_test, previsoes)
print(f"A acurácia foi {acuracia*100:.2f} %")
```

# Estratificando os casos de acordo com as classes

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25,random_state=20,stratify = y)

modelo = LinearSVC()
modelo.fit(X_train, y_train)
previsoes = modelo.predict(X_test)

acuracia = accuracy_score(y_test, previsoes)
print(f"A acurácia foi {acuracia*100:.2f} %")
```
